## 5. Train

`get_device()` auto-detects CUDA, so it uses the GPU with no code changes.
On a T4 this is roughly 15-20 minutes and reaches val loss ~1.5.

```python
!python -m chargpt.train \
  --steps 5000 --eval-interval 500 \
  --batch-size 64 --block-size 256 \
  --n-embd 384 --n-head 6 --n-layer 6 \
  --dropout 0.2 --lr 3e-4 \
  --out {OUT}
```

## 6. Generate a sample

```python
!python -m chargpt.sample --ckpt {OUT} --prompt "ROMEO:" --tokens 1000
```
